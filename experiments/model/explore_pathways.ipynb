```python
import reactome2py as reactome

def get_genes_in_pathway(pathway_name):
    # Search for the pathway by name
    pathways = reactome.content.search(pathway_name)
    
    # Check if the pathway exists
    if not pathways:
        raise ValueError(f"Pathway '{pathway_name}' not found.")
    
    # Retrieve the first matching pathway
    pathway = pathways[0]
    
    # Get the list of genes associated with the pathway
    genes = reactome.content.get_genes(pathway['stId'])
    
    return genes

# Example usage
pathway_name = 'EGFR Signaling'
genes = get_genes_in_pathway(pathway_name)
print(f"Genes in the '{pathway_name}' pathway:", genes)
```
